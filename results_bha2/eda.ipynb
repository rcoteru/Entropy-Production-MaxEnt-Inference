```python
# get all files in the given folder
import pathlib

folder = "dataset/iPA_183/ts/"

files = list(pathlib.Path(folder).glob("*.txt"))

# create a dataframe for each file
dfs = []
for file in files:
    df = pd.read_table(file, header=None)
    dfs.append(df)
```

```python
# plot the columns in the first dataframe

ndfs = 4

for i in range(ndfs):
    plt.plot(np.arange(i*len(dfs[i]), (i+1)*len(dfs[i])), dfs[i][0], label=f"File {i+1}")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Time Series Data from Multiple Files")
```
